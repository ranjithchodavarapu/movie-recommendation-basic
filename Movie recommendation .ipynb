```python
#import lib
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
data=pd.read_csv('movie_data.csv') #load the data
```

```python
data.head(3)   #return three rows
```

```python
data.shape
```

```python
columns=['Title','Genre','Actors','Director']   #main columns
```

```python
data[columns].head(3)
```

```python
#get missing data

data[columns].isnull().any()
```

```python
def important_features(data):
    important=[]
    for i in range(0,data.shape[0]):
        important.append(data['Actors'][i]+" "+data['Director'][i]+" "+data["Genre"][i]+" "+data["Title"][i])
    return important
```

```python
data["important"]=important_features(data)

data.head(3)
```

```python
# convert text to matrix
cm=CountVectorizer().fit_transform(data['important'])
```

```python
cs=cosine_similarity(cm)
print(cs)
```

```python
# shape of cs
cs.shape
```

```python
#get the title of the movie

title="Moana"

#find id
movie_id=data[data.Title==title]['Rank'].values[0]
```

```python
scores=list(enumerate(cs[movie_id]))
```

```python
sorted_score=sorted(scores,key=lambda x:x[1],reverse=True)

sorted_score=sorted_score[1:]
```

```python
print(sorted_score)
```

```python
# print the similar movies


j=0
for item in sorted_score:
    movie_title=data[data.Rank==item[0]]['Title'].values[0]
    print(j+1,movie_title)
    j=j+1
    if j>6:
        break
```
